# Single File (Decay) Analysis and plotting

```python
import numpy as np
import photonics.fluorescence as fl
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
# Create data
a = 1             # Amplitude
tau = 10          # Lifetime in ms
c = 0             # Noise offset
p_init = [a, tau, c]

x = np.linspace(0, 8*tau, num=1000)
y = fl.decay_fn(x, *p_init)

# Add noise to signal. Std of % a
y += np.random.normal(0, 0.1*a, 1000)

plt.figure()
plt.plot(x, y)
```

```python
# prepare data
y = fl.normalise(y)

plt.figure()
plt.plot(x, y)
```

```python
# Fit data
popt = fl.fit_decay(x, y)
popt
```

```python
# Plot results
fig = fl.plot_decay(x, y, fl.decay_fn, popt, log=True)
```
